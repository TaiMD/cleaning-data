```python
import json
import numpy as np
import pandas as pd
import io
import os
from google.colab import files
from tqdm import tqdm
```

```python
# Setting up lists for importing the data.
path_data = '/content/drive/MyDrive/Tai-Sana_exchangeDrive/pr-1_1/dataset'
list_of_folders = os.listdir(path_data)
list_of_folders
```

```python
list_of_directories = [path_data + '/' + i for i in list_of_folders]
list_of_directories
```

```python
list_of_dfs=[]
for i in tqdm(list_of_directories):
  list_of_csv = os.listdir(i)
  for j in list_of_csv:
      if j.endswith(".csv"):
        
        # Read csv
        k = pd.read_csv(i + '/' + j, sep='delimiter', engine ='python')
        
        k[["Column_1","MessageId","EventUtcTime","Timestamp","Cmd","SourceId","Payload","PayloadJson"]] = k['Column_1;MessageId;EventUtcTime;Timestamp;Cmd;SourceId;Payload;PayloadJson'].str.split(";", expand=True)
        k.drop(columns=['Column_1;MessageId;EventUtcTime;Timestamp;Cmd;SourceId;Payload;PayloadJson','Column_1','MessageId','EventUtcTime','Timestamp'], inplace=True)
        
        # We find the ACs that have HEARTBEAT cmds
        k = k.loc[k['Cmd'] == 'CMD_HEARTBEAT']
        
        k = k.replace(r'^\s*$', np.NaN, regex=True)
        k = k.replace(r'null', np.NaN, regex=True)
        

        # Payload and SourceID should not be empty
        k.dropna(subset=['SourceId','Payload'],inplace = True)
        

        # the Payload should not be '{}'
        k = k[(k.Payload != '{}')]
       
        k.reset_index(inplace = True, drop = True)
        # Expand json column to dataframe
        json1 = pd.json_normalize(k['Payload'].apply(json.loads))
        k2 = k.join(json1)
        k2 = k2.replace(r'^\s*$', np.NaN, regex=True)

        # At least one of the values of the payload should be != 255 and != NaN
        columns = ['fcuTcTemp','fcuTcjTemp','fcuFanRpm','cduTdTemp', 'cduTsTemp',
                   'cduTeTemp', 'cduCompHz','cduFanRpm', 'cduPmvPulse', 'cduIac']
                   
        k2 = k2[k2[columns].apply(lambda x:~x.isin(['ff', np.NaN])).any(axis=1)]
        k2 = k2['SourceId']

        # Return DataFrame with duplicate rows removed.
        k2.drop_duplicates(inplace = True)
        
        # storing k2 in list_of_dfs
        list_of_dfs.append(k2)
```

```python
# concatenating the list_of_dfs along with the vertical axis
df = pd.concat(list_of_dfs)
print(df)
# Convert df into a dataframe
df=df.to_frame()
df.reset_index(inplace=True)
df.drop(columns=['index'], inplace=True)

# Check duplicate rows in df
df.drop_duplicates(inplace = True)
print(len(df))

df_4500 = pd.read_csv('/content/drive/MyDrive/Tai-Sana_exchangeDrive/pr-1_1/newAcs.csv', names=['SourceId'])

df1 = pd.concat([df, df_4500], axis=0)
print(len(df1))

df1.drop_duplicates(inplace = True)
print(len(df1))
df1 = df1.reset_index()
df1.drop(columns=['index'], inplace = True)

df1.to_csv('/content/drive/MyDrive/Tai-Sana_exchangeDrive/pr-1_1/updated_newACs-0108.csv')
```
